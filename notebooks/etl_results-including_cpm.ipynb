## Generate total impressions, total clicks, and total conversions by country, channel_name, and business_vertical separately.

### 1. Total impressions, Total clicks, and Total conversions by `country`

```python
aggByCountry = jumboData.groupby(['country'])[['impressions', 'clicks', 'conversions', 'total_spend_cpm']].agg(sum).reset_index()
aggByCountry['im_cpm'] = aggByCountry['impressions'] / aggByCountry['total_spend_cpm']
aggByCountry['clicks_cpm'] = aggByCountry['clicks'] / aggByCountry['total_spend_cpm']
aggByCountry['conversions_cpm'] = aggByCountry['conversions'] / aggByCountry['total_spend_cpm']
aggByCountry['clicks_im'] = aggByCountry['clicks'] / aggByCountry['impressions']
aggByCountry['conversions_im'] = aggByCountry['conversions'] / aggByCountry['impressions']
aggByCountry['conversions_clicks'] = aggByCountry['conversions'] / aggByCountry['clicks']
aggByCountry
```

```python
aggByChannel = jumboData.groupby(['channel_name'])[['impressions', 'clicks', 'conversions', 'total_spend_cpm']].agg(sum).reset_index()
aggByChannel['im_cpm'] = aggByChannel['impressions'] / aggByChannel['total_spend_cpm']
aggByChannel['clicks_cpm'] = aggByChannel['clicks'] / aggByChannel['total_spend_cpm']
aggByChannel['conversions_cpm'] = aggByChannel['conversions'] / aggByChannel['total_spend_cpm']
aggByChannel['clicks_im'] = aggByChannel['clicks'] / aggByChannel['impressions']
aggByChannel['conversions_im'] = aggByChannel['conversions'] / aggByChannel['impressions']
aggByChannel['conversions_clicks'] = aggByChannel['conversions'] / aggByChannel['clicks']
aggByChannel
```

```python
aggByBusinessVertical = jumboData.groupby(['business_vertical'])[['impressions', 'clicks', 'conversions', 'total_spend_cpm']].agg(sum).reset_index()
aggByBusinessVertical['im_cpm'] = aggByBusinessVertical['impressions'] / aggByBusinessVertical['total_spend_cpm']
aggByBusinessVertical['clicks_cpm'] = aggByBusinessVertical['clicks'] / aggByBusinessVertical['total_spend_cpm']
aggByBusinessVertical['conversions_cpm'] = aggByBusinessVertical['conversions'] / aggByBusinessVertical['total_spend_cpm']
aggByBusinessVertical['clicks_im'] = aggByBusinessVertical['clicks'] / aggByBusinessVertical['impressions']
aggByBusinessVertical['conversions_im'] = aggByBusinessVertical['conversions'] / aggByBusinessVertical['impressions']
aggByBusinessVertical['conversions_clicks'] = aggByBusinessVertical['conversions'] / aggByBusinessVertical['clicks']
aggByBusinessVertical
```

## Select strategy_id =3718750. Show the cumulative sum of impressions and clicks partitioned by channel_name and region over date in ascending order. 

```python
specificStrategy = jumboData[ jumboData['strategy_id'] == 3718750 ]
```

```python
specificStrategy[:5]
```

```python
cumulativeSumDF = []
for index, df in specificStrategy.groupby(['channel_name', 'region']):
    df_sorted                      = df.sort_values('capturedDate').reset_index(drop=True)[['capturedDate', 'impressions', 'clicks', 'total_spend_cpm']]
    df_sorted['impressions_cumul'] = df_sorted['impressions'].cumsum()
    df_sorted['clicks_cumul']      = df_sorted['clicks'].cumsum()
    df_sorted['cpm_cumul']         = df_sorted['total_spend_cpm'].cumsum()
    df_sorted['channel_name']      = index[0]
    df_sorted['region']            = index[1]
    cumulativeSumDF.append( df_sorted )

cumulativeSumDF = pd.concat( cumulativeSumDF )
```
